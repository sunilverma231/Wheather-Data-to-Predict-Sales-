### Grouping and aggregating function 
Sometimes feeling cold is more then about the low temprature;a windy can also make you cold, a factor called the chill factor can be used to quantify the cold based on the wind speed and the temp. the formula for the chill factor is given by 
WCI = (10* sqrt(v)-v+10.5)*(33-mintemp)
v is the speed of the speed 
mintemp = minmum temprature 
add a column for WCI and find the month with the lowest WCI 


```python
# Creating the define function for WCI 
from math import sqrt 
def wci(x):
    velocity = x["WindGustSpeed"]
    minTemp = x["MinTemp"]
    return (10*sqrt(velocity)-velocity+10.5)*(33-minTemp)
```

```python
# Applying the function to the dataset 
w_df["WCI"] = w_df.apply(wci, axis = 1)
w_df.head()
```
